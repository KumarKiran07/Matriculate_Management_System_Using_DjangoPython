```python
import pandas as pd
```

```python
chi2 = pd.read_csv('chisquare1.csv')
chi2
```

```python
contingency = pd.crosstab(chi2['Buyer'], chi2['age'])
contingency
```

```python
alpha = 1.0
from scipy.stats import chi2_contingency
chi2, p , dof, e = chi2_contingency(contingency)
if p > alpha:
    print('reject_H0')
else:
    print('accept_H0')
```

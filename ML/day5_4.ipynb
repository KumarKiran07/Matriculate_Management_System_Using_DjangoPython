```python
import pandas as pd 
temp = pd.read_csv('test1.csv')
temp
```

```python
contingency = pd.crosstab(temp['Gender'], temp['Education'])
contingency
```

```python
alpha = 1.0
from scipy.stats import chi2_contingency
chi2, p , dof, e = chi2_contingency(contingency)
if p > alpha:
    print('reject_H0')
else:
    print('accept_H0')
```

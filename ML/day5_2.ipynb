```python
import pandas as pd
```

```python
chi3 = pd.read_csv('athelete.csv')
chi3
```

```python
contingency = pd.crosstab(chi3['class rank'], chi3['gender'])
contingency
```

```python
alpha = 1.0
from scipy.stats import chi2_contingency
chi2, p , dof, e = chi2_contingency(contingency)
if p > alpha:
    print('reject_H0')
else:
    print('accept_H0')
```

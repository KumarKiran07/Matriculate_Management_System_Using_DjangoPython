```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
student = pd.read_csv('year_student.csv')
#initialize the list for x and y 
x = list(student.iloc[:,0])
y = list(student.iloc[:,1])

plt.bar(x, y, color ='gray')
plt.title('Student over 12 year')
plt.xlabel('Years')
plt.ylabel('Number of student ')
plt.show()
```

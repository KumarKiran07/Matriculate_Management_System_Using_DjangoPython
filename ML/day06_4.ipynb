```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
temp = pd.read_csv('insurance.csv',usecols=['age','bmi'])
temp
```

```python
plt.xlabel('age')
plt.ylabel('bmi')
plt.scatter(temp.age, temp.bmi, color='red', marker='*')
```

```python
new_temp = temp.drop('bmi', axis='columns')
new_temp
```

```python
model = linear_model.LinearRegression()
model.fit(new_temp,temp.bmi)
```

```python
model.predict([[120]])
```
